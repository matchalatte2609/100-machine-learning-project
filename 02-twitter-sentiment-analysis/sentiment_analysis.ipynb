```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```python
# Load Sentiment140 dataset from Kaggle
df = pd.read_csv('training.1600000.processed.noemoticon.csv', encoding='latin-1', header=None)
df = df[[0, 5]]
df.columns = ['polarity', 'text']
print(df.head())
```

## EDA

```python
# We remove neutral tweets where polarity is 2, maps the labels so 0 = negative, 4 -> 1 = positive

df = df[df.polarity != 2]
df['polarity'] = df['polarity'].map({0: 0, 4: 1})

print(df['polarity'].value_counts())
```

The data seems balanced. Great!

```python
# visualize class distribution
plt.figure(figsize=(8, 5))
df['polarity'].value_counts().plot(kind='bar', color=['#e74c3c', '#2ecc71'])
plt.title('Class Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks([0, 1], ['Negative', 'Positive'], rotation=0)
plt.tight_layout()
plt.show()
```

```python
#clean the tweets, convert all text to lowercase
def clean_text(text):
    return text.lower()
df['clean_text'] = df['text'].apply(clean_text)
print(df[['text', 'clean_text']].head())
```

works just fine

## Splitting Data

```python
X_train, X_test, y_train, y_test = train_test_split(
    df['clean_text'],
    df['polarity'],
    test_size=0.2,
    random_state=42
)

print("Train size:", len(X_train))
print("Test size:", len(X_test))
```

```python
# Creating TF IDF vectorizer that converts text -> num features using unigrams and bigrams, limited to 5k features
vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))

# fit and transofrm data
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

print("TF-IDF shape (train):", X_train_tfidf.shape)
print("TF-IDF shape (test):", X_test_tfidf.shape)
```

## Model Training

```python
#Bernoulli Naive Bayes
bnb = BernoulliNB()
bnb.fit(X_train_tfidf, y_train)
bnb_pred = bnb.predict(X_test_tfidf)
print("Bernoulli Naive Bayes Accuracy:", accuracy_score(y_test, bnb_pred))
print("\nBernoulliNB Classification Report:\n", classification_report(y_test, bnb_pred))
```

```python
# confusion matrix for BernoulliNB
cm_bnb = confusion_matrix(y_test, bnb_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm_bnb, annot=True, fmt='d', cmap='Blues', xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'])
plt.title('Confusion Matrix - BernoulliNB')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()
plt.show()
```

```python
# Supported Vector Machine
svm = LinearSVC(max_iter=1000)
svm.fit(X_train_tfidf, y_train)
svm_pred = svm.predict(X_test_tfidf)

print("SVM Accuracy:", accuracy_score(y_test, svm_pred))
print("\nSVM Classification Report:\n", classification_report(y_test, svm_pred))
```

```python
# confusion matrix for SVM
cm_svm = confusion_matrix(y_test, svm_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm_svm, annot=True, fmt='d', cmap='Greens', xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'])
plt.title('Confusion Matrix - SVM')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()
plt.show()
```

```python
# LR
logreg = LogisticRegression(max_iter=100)
logreg.fit(X_train_tfidf, y_train)
logreg_pred = logreg.predict(X_test_tfidf)

print("Logistic Regression Accuracy:", accuracy_score(y_test, logreg_pred))
print("\nLogistic Regression Classification Report:\n", classification_report(y_test, logreg_pred))
```

```python
# confusion matrix for Logistic Regression
cm_logreg = confusion_matrix(y_test, logreg_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm_logreg, annot=True, fmt='d', cmap='Oranges', xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'])
plt.title('Confusion Matrix - Logistic Regression')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()
plt.show()
```

```python
# test on some examples
test_tweets = [
    "I absolutely love this product! Best purchase ever!",
    "This is terrible, waste of money",
    "Just had the best day at the beach with friends",
    "Feeling really frustrated with customer service",
    "Not sure how I feel about this"
]

print("Sample Predictions:\n")
for tweet in test_tweets:
    cleaned = clean_text(tweet)
    vec = vectorizer.transform([cleaned])
    
    # get predictions from best model (Logistic Regression)
    pred = logreg.predict(vec)[0]
    prob = logreg.decision_function(vec)[0]
    
    sentiment = "Positive" if pred == 1 else "Negative"
    confidence = abs(prob)
    
    print(f"Tweet: {tweet}")
    print(f"Prediction: {sentiment} (confidence: {confidence:.2f})")
    print()
```

```python
# summary table
results_df = pd.DataFrame({
    'Model': ['BernoulliNB', 'SVM', 'Logistic Regression'],
    'Accuracy': [0.7665, 0.7953, 0.7961],
    'Precision (Neg)': [0.77, 0.80, 0.80],
    'Precision (Pos)': [0.76, 0.79, 0.79],
    'Recall (Neg)': [0.75, 0.78, 0.78],
    'Recall (Pos)': [0.78, 0.81, 0.81],
    'F1-Score': [0.77, 0.80, 0.80]
})

print("Model Performance Summary:")
print(results_df.to_string(index=False))
```

```python
# compare model accuracies
models = ['BernoulliNB', 'SVM', 'Logistic Regression']
accuracies = [
    accuracy_score(y_test, bnb_pred),
    accuracy_score(y_test, svm_pred),
    accuracy_score(y_test, logreg_pred)
]

plt.figure(figsize=(10, 6))
bars = plt.bar(models, accuracies, color=['#3498db', '#2ecc71', '#e67e22'])
plt.ylim(0.7, 0.85)
plt.title('Model Performance Comparison')
plt.ylabel('Accuracy')
plt.xlabel('Model')

# add values on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.4f}',
             ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

## Key Takeaways

Based on the results:

- **Best Model**: Logistic Regression achieved the highest accuracy at 79.6%, closely followed by SVM at 79.5%
- **Scale**: The model can classify 320,000 tweets with ~80% accuracy, which translates to correctly identifying sentiment in roughly 255,000 tweets
- **Practical Use**: This level of performance is suitable for basic sentiment monitoring tasks like tracking brand mentions or general public opinion trends
- **Class Balance**: Both models perform equally well on positive and negative sentiments (81% recall for positive, 78% for negative)
- **Speed vs Accuracy**: BernoulliNB is faster but sacrifices ~3% accuracy compared to Logistic Regression

This approach works well for basic sentiment analysis but could be improved with better text preprocessing and more advanced models.
